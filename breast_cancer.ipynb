```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(NEW_X,Y, test_size=0.20, random_state=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adadelta, Adam, Adagrad, Adamax, Nadam, RMSprop 
from tensorflow.keras.losses import mse
```

```python
#SGD
# Use the required optimizer to compile the model
# The number of hidden neurons should be between the size of the input layer and the size of the output layer.
# The number of hidden neurons should be 2/3 the size of the input layer, plus the size of the output layer.
# The number of hidden neurons should be less than twice the size of the input layer.
# (2/3*10 + 2 = 7)
def SGD_model(x_train, y_train, x_test, y_test,ax):
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # model.add(Dense(1))

    print("Printing behins here: ----------------")
    model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.05,momentum=0.0,
        nesterov=False,),loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'], label='loss')
    ax[0,0].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    ax[0,0].set_title('SGD')
    # ax[0,0].set_title('SGD')
    ax[0,0].set_xlabel('Epochs')
    ax[0,0].set_ylabel('Loss')
    # plt.ylabel('loss')
    pred = model.predict(x_test)
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
    # model.predict(x_test,y_test)
```

```python
def MINI_BATCH_SGD_model(x_train, y_train, x_test, y_test,ax):
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # model.add(Ac)
    print("Printing behins here: ----------------")
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.05,momentum=0.0,
        nesterov=False),loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100, batch_size=20)

    #loss over time
    # plt.plot(train.history['loss'] ,label='loss')
    ax[0,1].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[0,1].title.set_text('Mini Batch SGD')
    ax[0,1].set_xlabel('Epochs')
    ax[0,1].set_ylabel('Loss')
    pred = model.predict(x_test)
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```

```python
def MOMENTUM_model(x_train, y_train, x_test, y_test,ax):
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # model.add(Ac)
    print("Printing behins here: ----------------")
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.05,momentum=0.9,
        nesterov=False),loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'] ,label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[1,0].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[1,0].title.set_text('Momentum')
    ax[1,0].set_xlabel('Epochs')
    ax[1,0].set_ylabel('Loss')
    pred = model.predict(x_test)
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```

```python
def NESTEROV_model(x_train, y_train, x_test, y_test,ax):
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # model.add(Ac)
    print("Printing behins here: ----------------")
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.05,momentum=0.9,
        nesterov=True),loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'] ,label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[1,1].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[1,1].title.set_text('Nesterov')
    ax[1,1].set_xlabel('Epochs')
    ax[1,1].set_ylabel('Loss')
    pred = model.predict(x_test)
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```

```python
def RMSPROP_model(x_train, y_train, x_test, y_test,ax):
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # model.add(Ac)
    print("Printing behins here: ----------------")
    model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.05, rho=0.9, epsilon=1e-8) ,loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'] ,label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[2,0].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[2,0].title.set_text('Rmsprop')
    ax[2,0].set_xlabel('Epochs')
    ax[2,0].set_ylabel('Loss')
    pred = model.predict(x_test)
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```

```python
def ADAM_model(x_train, y_train, x_test, y_test,ax):
    #Creating Sequential Model
    # lr_schedule = keras.optimizers.schedules.l
    print("Printing behins here: ----------------")
    model = Sequential()
    # model.add(Dense(1))
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # Use the required optimizer to compile the model
    model.compile(optimizer=Adam(learning_rate=0.05,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-8,
        amsgrad=False,),loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'],label='loss')
    
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[2,1].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[2,1].title.set_text('Adam')
    ax[2,1].set_xlabel('Epochs')
    ax[2,1].set_ylabel('Loss')
    pred = model.predict(x_test)
    
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```

```python
def ADAGRAD_model(x_train, y_train, x_test, y_test,ax):
    #Creating Sequential Model
    print("Printing behins here: ----------------")
    model = Sequential()
    model.add(Dense(10,activation = 'sigmoid'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(2,activation = 'sigmoid'))
    # Use the required optimizer to compile the model
    model.compile(optimizer=Adagrad(
        learning_rate=0.05,
        initial_accumulator_value=0.1,
        epsilon=1e-7),loss=tf.keras.losses.categorical_crossentropy)
    train = model.fit(x_train,y_train,epochs=100)

    #loss over time
    # plt.plot(train.history['loss'],label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[3,0].plot(train.history['loss'], label='loss')
    # plt.xlabel('epochs')
    # plt.ylabel('loss')
    ax[3,0].title.set_text('Adagrad')
    ax[3,0].set_xlabel('Epochs')
    ax[3,0].set_ylabel('Loss')
    pred = model.predict(x_test)
    
    return accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
```
